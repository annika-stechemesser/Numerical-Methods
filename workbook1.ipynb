We calculate the first 80 terms in the sequence defined above once in single precision floating point arithmetic and one in double precision floating point arithmetic. The results are plotted below.

```julia
N=80
rec32=Array{Float32}(N)
rec32[1]=1
rec32[2]=2/3

for i=3:N
    rec32[i]=2*rec32[i-1]-(8/9)*rec32[i-2]
end


N=80
rec64=Array{Float64}(N)
rec64[1]=1
rec64[2]=2/3

for i=3:N
    rec64[i]=2*rec64[i-1]-(8/9)*rec64[i-2]
end

x=1:1:80
plot(x,rec32, yscale=:log10, linewidth=3, label=L"$a_n$ 32bits",title=L"Numerical solution of the recursion $a_n$ [lin/log plot]")
plot!(x,rec64, yscale=:log10, line=(:dash,3), linewidth=2, label=L"$a_n$ 64bits",xlabel="n")
```

#### Question 1 Part 3

We want to find a general solution to $a_{n+1} = 2\, a_n - \frac{8}{9}\, a_{n-1}$

Ansatz: $a_n = \lambda^n$

We then get:


\begin{align}
&\lambda^{n+1} = 2 \lambda^n - \frac{8}{9} \lambda^{n-1} \\
&\Leftrightarrow \lambda^{n+1} - 2 \lambda^n + \frac{8}{9} \lambda^{n-1} = 0 \\
&\Leftrightarrow \lambda^{n-1}(\lambda^2 - 2 \lambda + \frac{8}{9} 1) = 0 \\
&\Leftrightarrow \lambda = 0 \lor \lambda = 1 \pm \sqrt{1-\frac{8}{9}} = 1 \pm \frac{1}{3} \\
&\Rightarrow \lambda_3 = 0, \lambda_2 = \frac{2}{3}, \lambda_1 = \frac{4}{3} \\
&\Rightarrow a_{n}=C_1\lambda_1^n+C_2\lambda_2^n \\
&\Rightarrow a_{n}=C_1(\frac{4}{3})^n+C_2(\frac{2}{3})^n \\
\end{align}

#### Question 1 Part 4

Using the inital conditions $a_1=1$ and $a_2=\frac{2}{3}$ we get :

$$ 1 = \frac{4}{3}C_1+ \frac{2}{3}C_2 $$ and 
$$ \frac{2}{3} = \frac{16}{9} C_1 + \frac{4}{9}C_2$$

Solving this gives $C_1 = 0 $ and $C_2 = \frac{3}{2}$

#### Question 1 Part 5

Now we consider the perturbated initial conditions $a_1=1$ and $a_2=\frac{2}{3}+\epsilon$. This gives the equations

$$ 1 = \frac{4}{3}C_1^*+ \frac{2}{3}C_2^*$$ and 

$$ \frac{2}{3}+\epsilon = \frac{16}{9} C_1^* + \frac{4}{9}C_2^*$$

Solving this gives $C_1^* = \frac{9}{8}\epsilon $ and $C_2^* = \frac{3}{2}(1-\frac{3}{2}\epsilon)$.

We can use this solution to explain our numeric results. Before $C_1$ was equal to 0. Now $C_1$ is very small but not equal to zero. The solution to the recursion is $a_{n}=C_1(\frac{4}{3})^n+C_2(\frac{2}{3})^n$. If $C_1$ is even slightly bigger than 0, the first term will take over for some power of n (as 4/3 is bigger than 2/3).  If we consider single precision floating point arithmetics we have less bits to accurately express our floats and have therefore a higher perturbation. The first term takes over relatively fast ( circa n=25 ). In double precision floating point arithmetics we can express the floats more accurately but we still get a (smaller) perturbation. The first term takes over at n $\sim$ 58. To get a solution that matches the theoretical prediction, we use the BigFloat type.
            

#### Question 1 Part 6
We now use the BigFloat type to reproduce the exact solution for the first 80 terms.   

```julia
N=80
rec128=Array{BigFloat}(N)
rec128[1]=BigFloat(1)
rec128[2]=BigFloat(2)/BigFloat(3)

for i=3:N
    rec128[i]=BigFloat(2)*BigFloat(rec128[i-1])-(BigFloat(8)/BigFloat(9))*BigFloat(rec128[i-2])
end

#analytical solution
rec= recursion_theo(N) #NUMBER 1.1 IN FILE


x=1:1:80
plot(x,rec128, yscale=:log10, linewidth=6,label=L"$a_n$ 128bits",xlabel="n")
scatter!(x,rec, yscale=:log10, linewidth=2, label="Analytical")
```

We see that the solution using arbitrary precision floating point arithmetic is for the first 80 iterations identical with the analytical solution. 

## Question 2: Computational complexity of the mergesort algorithm

Consider two arrays of integers, A and B, having lengths n and m respectively. Assuming that the elements of A and B are already sorted in ascending order. The following recursive function merges them to return an array of length n+m whose elements are sorted in ascending order:



```julia
function mergepresorted(A::Array{Int64,1}, B::Array{Int64,1})
    if length(A) == 0
        return B
    elseif length(B) == 0
        return A
    elseif A[1] < B[1]
        return vcat([A[1]], mergepresorted(A[2:end], B))
    else
        return vcat([B[1]], mergepresorted(A, B[2:end]))
    end    
end
```

We now write a recursive function that implements the mergesort algorithm. We assume that the input array is of size $n=2^m$.
```julia
## use the helper function mergepresorted

function my_mergesort(A)   #NUMBER 2.2 IN FILE
    n=length(A)
    if n == 1
        return A 
    else m=n/2
        m=Int(m)
        return mergepresorted(my_mergesort(A[1:m]),my_mergesort(A[m+1:n]))
    end    
end 
```

#### Question 2 Part 3
The computational complexity of the merge sort is $F(n) = 2 F(\frac{n}{2})+n$ because if we have an array of size n, we split the it into two subproblems. Each of the subarrays has length $\frac{n}{2}$.To solve each of these problems, we need a computational effort of $F(\frac{n}{2})$, therefore we get the term $2 F(\frac{n}{2})$. In addition to that we have to consider the fact that we need to apply the "helper" function (mergepresorted) to the subproblems to get a solution in the end. As stated above, this has a computational effort of $n+m$ where $n$ is the length of array $A$ and $m$ is the length of array $B$. As both subarrays are here of size $\frac{n}{2}$, the computational effort of mergepresorted is n. So, in total, $F(n)$ needs to be $2 F(\frac{n}{2})+n$. 

#### Question 2 Part 4
Let $n=2^p$ and let $b_p=F(2^p)$. We can the express the recurrence relation above in terms of $b_p$:

$F(n) = 2 F(\frac{n}{2})+n  \,\,, F(1)=1 \,\,\,\, \Rightarrow F(2^p)=2F(2^{p-1})+2^p  \,\,, F(2^0)=1 \,\,\,\, \Rightarrow b_p=2b_{p-1}+2^p  \,\,, b_0=1$

#### Question 2 Part 5
First we want to solve the homogeneous problem $b_p=2b_{p-1}$.

\begin{align}
b_p&=2b_{p-1} \\
Ansatz: b_p&=x^p \\
x^p&=2x^{p-1} \\
\Rightarrow x&=2 \\ 
\end{align}

#### Question 2 Part 6

Now we want to find a particular solution of the inhomogeneous problem. 

Ansatz: $b_p = pA2^p+B$

Plugging this into the formula for $b_p$ we obtained above we get 

\begin{align}
pA2^p+B&=2((p-1)A2^{p-1}+B)+2^p \\
pA2^p+B&=2^p(pA-A+1)+2B \\
 pA&=pA-A+1 \,\, \\
\Rightarrow A=1 \,\,\,\, &and \,\, 2B=B \Rightarrow B=0 \\
\end{align}
So overall we get $b_p=C2^p+p2^p$

We use the initial condition $b_0=1$ to determine the value of the constant $C$:

$$b_0 = 1 \,\, \Rightarrow \,\, 1=C2^0+0 \,\, \Rightarrow C=1$$

So finally, $b_p=2^p(1+p)$

#### Question 2 Part 7

We now want to compute the computational complexity of the mergesort algorithm. Remember that $n=2^p$. Therefore $p=\frac{log(n)}{log(2)}$. We plug this in for p and get 

\begin{align}
F(n)&= b_{\frac{log(n)}{log(2)}}=2^{\frac{log(n)}{log(2)}}(1+\frac{log(n)}{log(2)})\\
F(n)&= (2^{log(n)})^{\frac{1}{log(2)}}(1+\frac{log(n)}{log(2)})\\
F(n)&= (n^{log(2)})^{\frac{1}{log(2)}}(1+\frac{log(n)}{log(2)})\\
F(n)&= n(1+\frac{log(n)}{log(2)}) \,\,\sim \,\, \mathcal{O}(n\log{}n) \\
\end{align}
